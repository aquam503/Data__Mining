### Feature Selection

```python
df.head()
```

```python
df.drop(["Unnamed: 0","listing_id","host_id","host_name","last_review","name","neighbourhood","listing_added"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df["borough"].unique()
```

```python
mapping={'Brooklyn':0,
         'Manhattan':1,
         'Bronx':2,
         'Queens':3,
         'Staten Island':4}
df["borough"]=df["borough"].replace(mapping)
mapping={'entire place':0,
         'Private Room':1,
         'Shared Room':2
        }
df["room_type"]=df["room_type"].replace(mapping)
```

```python
df["borough"].unique()
```

```python
df["room_type"].unique()
```

```python
df.head()
```

```python
df_train=df.dropna(subset=["rating"])
```

```python
df_train.isna().sum()
```

```python
df_train["class"]=np.where(df_train["rating"]>3.5,1,0)
```

```python
df_train["class"].unique()
```

```python
df_train.isna().sum()
```

```python
df_train.head()
```

```python
df_test=df[df["rating"].isnull()]
```

```python
X_train=df_train.drop(["is_rated","rating","class"],axis=1)
y_train=df_train["class"]
```

```python
X_train.columns
```

```python
len(X_train.columns)
```

```python
X_train.isna().sum()
```

```python
y_train.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_train)
```

```python
X_scaled=pd.DataFrame(X_scaled,columns=X_train.columns)
```

```python
X_scaled.isna().sum()
```

```python
# Apply the SelectKBest class to extract top 10 best features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
best_features = SelectKBest(score_func=chi2, k=6)
fit = best_features.fit(X_scaled, y_train)

# Get the selected feature names and scores
feature_scores = pd.DataFrame(list(zip(X_scaled.columns, fit.scores_)), columns=['Feature', 'Score'])

# Sort the features by score in descending order
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Print the selected features and their scores
print("Selected Features:")
feature_scores.head(10)
```

```python
columns=list(feature_scores["Feature"].values)
```

```python
columns
```

```python
df_selected=X_scaled[columns]
```

```python
df_selected.isna().sum()
```

```python
y_train.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_selected)

# plot scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
pca=PCA(n_components=5)
X_pca=pca.fit_transform(df_selected)
# Print the explained variance ratio for each component
print('Explained variance ratio:', pca.explained_variance_ratio_)

# Print the total explained variance ratio for all components
print('Total explained variance ratio:', sum(pca.explained_variance_ratio_))
```

```python
pca_columns = ['PCA'+str(i+1) for i in range(5)]
df_pca = pd.DataFrame(X_pca, columns=pca_columns)
```

```python
df_pca["class"]=list(y_train)
```

```python
df_pca.to_csv("airbnb_selected.csv",index=False)
```

### modeling

```python
#### splitting data
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
```

```python
X=X_train
y=y_train
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
### instantiating of model
models={"LinearSVC":LinearSVC(),
        "KNeighborsClassifier":KNeighborsClassifier(),
        "SVC":SVC(),
        "LogisticRegression":LogisticRegression(),
        "DecisionTreeClassifier":DecisionTreeClassifier(),
        "RandomForestClassifier":RandomForestClassifier()
        }
```

```python
results={}  
np.random.seed(42)
for model_name,model in models.items():
    model.fit(X_train,y_train)
    results[model_name]=model.score(X_test,y_test)

results
```

```python
results_df=pd.DataFrame(results.values(),results.keys(),columns=["Accuracy"])
```
